# Simple ACT Test on MNIST

This notebook tests the Adaptive Computation Time (ACT) mechanism on MNIST digit classification.

## Why MNIST?
- **Simple task**: Allows us to focus on ACT mechanics, not task complexity
- **Fast training**: 5-10 minutes instead of hours
- **Clear patterns**: Different digits should need different computation depths
- **Easy debugging**: Can visualize and understand everything

## Expected Behavior
- Simple digits (1, 7) should halt early (1-2 iterations)
- Complex digits (8, 9) should need more iterations (3-4)
- The model should learn this pattern automatically through RL
